```python
subset1 = df.iloc[indices + 1]
subset2 = df.iloc[indices + 2]
```

```python
colleges = []
stages = []
cutoffs = []

for (idx1, row1), (idx2, row2) in zip(subset1.iterrows(), subset2.iterrows()):
    for idx_row, item in row1.items():
        row0 = df.iloc[idx1 - 1, 0]
        if (item == "GOPENH" or item == "GOPENS") and "Savitribai Phule" in row0:
            colleges.append(row0)
            stages.append(item)
            cutoffs.append(row2.iloc[idx_row])
```

```python
idx1, idx2
```

```python
new_df = pd.DataFrame()
```

```python
new_df['stage'] = stages
new_df['cutoff'] = cutoffs
new_df['college'] = colleges
```
